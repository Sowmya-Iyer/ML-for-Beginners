<h2> Save And Load a Trained Model</h2>

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

## Pickle

```python
import pickle
```

Saving model using pickle.dump

here 'wb' represents write mode

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

Loading model using pickle.Loading

here 'rb' represents read mode

```python
with open('model_pickle','rb') as file:
    mp = pickle.load(file)
```

```python
mp.coef_
```

```python
mp.intercept_
```

```python
mp.predict([[5000]])
```

## Joblib from sklearn
Joblib works efficiently with large number of numpy arrays used

```python
from sklearn.externals import joblib
```

Saving model using joblib.dump 

```python
joblib.dump(model, 'model_joblib')
```

Loading model using joblib.load

```python
mj = joblib.load('model_joblib')
```

```python
mj.coef_
```

```python
mj.intercept_
```

```python
mj.predict([[5000]])
```
